```python
import os
from PyPDF2 import PdfReader
from softtek_llm.chatbot import Chatbot
from softtek_llm.models import OpenAI
from softtek_llm.cache import Cache
from softtek_llm.vectorStores import PineconeVectorStore
from softtek_llm.embeddings import OpenAIEmbeddings
from softtek_llm.schemas import Filter
from dotenv import load_dotenv
#from Los_zips.PruebaSecundaria import Main
```

```python
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if OPENAI_API_KEY is None:
    raise ValueError("OPENAI_API_KEY not found in .env file")

OPENAI_API_BASE = os.getenv("OPENAI_API_BASE")
if OPENAI_API_BASE is None:
    raise ValueError("OPENAI_API_BASE not found in .env file")

OPENAI_EMBEDDINGS_MODEL_NAME = os.getenv("OPENAI_EMBEDDINGS_MODEL_NAME")
if OPENAI_EMBEDDINGS_MODEL_NAME is None:
    raise ValueError("OPENAI_EMBEDDINGS_MODEL_NAME not found in .env file")

OPENAI_CHAT_MODEL_NAME = os.getenv("OPENAI_CHAT_MODEL_NAME")
if OPENAI_CHAT_MODEL_NAME is None:
    raise ValueError("OPENAI_CHAT_MODEL_NAME not found in .env file")
PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")
if PINECONE_API_KEY is None:
    raise ValueError("PINECONE_API_KEY not found in .env file")

PINECONE_ENVIRONMENT = os.getenv("PINECONE_ENVIRONMENT")
if PINECONE_ENVIRONMENT is None:
    raise ValueError("PINECONE_ENVIRONMENT not found in .env file")

PINECONE_INDEX_NAME = os.getenv("PINECONE_INDEX_NAME")
if PINECONE_INDEX_NAME is None:
    raise ValueError("PINECONE_INDEX_NAME not found in .env file")
```

```python
vector_store = PineconeVectorStore(
    api_key=PINECONE_API_KEY,
    environment=PINECONE_ENVIRONMENT,
    index_name=PINECONE_INDEX_NAME,
)
embeddings_model = OpenAIEmbeddings(
    api_key=OPENAI_API_KEY,
    model_name=OPENAI_EMBEDDINGS_MODEL_NAME,
    api_type="azure",
    api_base=OPENAI_API_BASE,
)
cache = Cache(
    vector_store=vector_store,
    embeddings_model=embeddings_model,
)
model = OpenAI(
    api_key=OPENAI_API_KEY,
    model_name=OPENAI_CHAT_MODEL_NAME,
    api_type="azure",
    api_base=OPENAI_API_BASE,
    verbose=False,
)
#filters = [
#    Filter(
#    )
#]
chatbot = Chatbot(
    model=model,
    description="You are a very helpful and polite chatbot",
    #filters=filters,
    cache=cache,
    verbose=False,
)
with open('Resumen.txt', 'w') as f:
        f.close()
with open('ResumeFinal.txt', 'w') as f:
        f.close()

chatbot.cache.vector_store.delete(delete_all=True)
```

```python
def Summary(resumen):
    response = chatbot.chat(
    prompt=f"Summary of the following text:\n{resumen}\n\Summary:",
    print_cache_score=True
    )
    print(response)
    dead = str(response)
    dead = dead[dead.find("content=")+9:dead.find(")")]
    with open('Summary.txt', 'w') as fd:
        fd.write(dead)
        fd.close()
    chatbot.cache.vector_store.delete(delete_all=True)


def KeyWords(resumen):
    response = chatbot.chat(
    prompt=f"Key words of the following text:\n{resumen}\n\Key words:",
    print_cache_score=True
    )
    print(response)
    dead = str(response)
    dead = dead[dead.find("content=")+9:dead.find(")")]
    with open('KeyWords.txt', 'w') as fd:
        fd.write(dead)
        fd.close()
    chatbot.cache.vector_store.delete(delete_all=True)
    

def WordSearch(resumen):
    keyWordSearch = input("Enter the word you want to search: ")
    response = chatbot.chat(
    prompt=f"search information of this keyword: {keyWordSearch} in the following text:\n{resumen}\n\Information of the word:",
    print_cache_score=True
    )
    dead = str(response)
    dead = dead[dead.find("content=")+9:dead.find(")")]
    print(response)
    with open('WordSearch.txt', 'w') as fd:
        fd.write(dead)
        fd.close()
    chatbot.cache.vector_store.delete(delete_all=True)

def QA(resumen):
    Question = input("Enter your question about the text: ")
    response = chatbot.chat(
    prompt=f"Answer the next question: {Question} about the following text:\n{resumen}\n\Answer:",
    print_cache_score=True
    )
    dead = str(response)
    dead = dead[dead.find("content=")+9:dead.find(")")]
    print(response)
    with open('QA.txt', 'w') as fd:
        fd.write(dead)
        fd.close()
    chatbot.cache.vector_store.delete(delete_all=True)

#una funcion que activa el resumen y guarda el output en un .txt para su uso futuro.
def BigWordsShort(texto_a_resumir):
    response = chatbot.chat(
    prompt=f"Summary the following text to the most important only, less tan 1000 words\n{texto_a_resumir}\n\Summary:",
    print_cache_score=True
    )
    print(response)
    with open('Resumen.txt', 'a') as f:
        #convertir response a string y guardarlo en el txt
        dead = str(response)
        #en el string dead empezar a contar desde el caracter content=' hasta ') y guardarlo en el txt
        dead = dead[dead.find("content=")+9:dead.find(")")]
        f.write(dead)
        #añadir un salto de linea
        f.write("\n")
        f.close()
    print("Resumen guardado en Resumen.txt")
    chatbot.cache.vector_store.delete(delete_all=True)

def BigWordsShort2(texto_a_resumir):
    response = chatbot.chat(
    prompt=f"Summary the following text to the most important\n{texto_a_resumir}\n\Summary:",
    print_cache_score=True
    )
    print(response)
    with open('ResumeFinal.txt', 'a') as f:
        #guarda en resumenfinal
        #convertir response a string y guardarlo en el txt
        dead = str(response)
        #en el string dead empezar a contar desde el caracter content=' hasta ') y guardarlo en el txt
        dead = dead[dead.find("content=")+9:dead.find(")")]
        f.write(dead)
        #añadir un salto de linea
        f.write("\n")
        f.close()
    print("Resumen guardado en Resumen.txt")
    chatbot.cache.vector_store.delete(delete_all=True)
    
    
    
```
